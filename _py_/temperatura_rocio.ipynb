```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
data = pd.read_csv('GCLP_All.csv')

# Convertir columnas de fecha
data['dateTime'] = pd.to_datetime(data['dateTime'])

# Agregar una columna de mes
data['Month'] = data['dateTime'].dt.to_period('M')

# Calcular promedios mensuales para temperatura y punto de roco
monthly_avg = data.groupby('Month').agg({'temperature': 'mean', 'dewPoint': 'mean'}).reset_index()

# Grfico de lneas para temperatura y punto de roco
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg['Month'].astype(str), monthly_avg['temperature'], label='Temperatura Promedio (C)', marker='o')
plt.plot(monthly_avg['Month'].astype(str), monthly_avg['dewPoint'], label='Punto de Roco Promedio (C)', marker='o')
plt.title('Promedio Mensual de Temperatura y Punto de Roco')
plt.xlabel('Mes')
plt.ylabel('Temperatura y punto de roco (C)')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
