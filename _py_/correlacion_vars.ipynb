```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos
data = pd.read_csv('GCLP_All.csv')

# Seleccionar las columnas relevantes
relevant_columns = ['temperature', 'dewPoint', 'qnh', 'visibility', 'knots']
correlation_data = data[relevant_columns].dropna()  # Eliminar filas con valores nulos

# Calcular la matriz de correlacin
correlation_matrix = correlation_data.corr()

# Crear el mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Mapa de calor de correlacin entre variables meteorolgicas')
plt.show()
```
