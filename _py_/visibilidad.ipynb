```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
data = pd.read_csv('GCLP_All.csv')

# Convertir columnas de fecha
data['dateTime'] = pd.to_datetime(data['dateTime'])

# Extraer el ao y mes de cada fecha y crear una nueva columna 'YearMonth'
data['YearMonth'] = data['dateTime'].dt.to_period('M').astype(str)

# Calcular el promedio mensual de visibilidad
monthly_visibility = data.groupby('YearMonth')['visibility'].mean().reset_index()

# Crear un grfico de lneas para mostrar los promedios mensuales
plt.figure(figsize=(14, 7))
plt.plot(monthly_visibility['YearMonth'], monthly_visibility['visibility'], marker='o', linestyle='-', color='royalblue')
plt.title('Cambios en la Visibilidad Media Mensual')
plt.xlabel('Mes (Ao-Mes)')
plt.ylabel('Visibilidad Media (metros)')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mejor legibilidad
plt.grid(True)
plt.tight_layout()  # Ajustar automticamente los parmetros del subplot para dar espacio a las etiquetas
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
data = pd.read_csv('GCLP_All.csv')

# Convertir columnas de fecha
data['dateTime'] = pd.to_datetime(data['dateTime'])

# Extraer el ao y mes de cada fecha y crear una nueva columna 'YearMonth'
data['YearMonth'] = data['dateTime'].dt.to_period('M').astype(str)

# Calcular el promedio mensual de visibilidad
monthly_visibility = data.groupby('YearMonth')['visibility'].mean().reset_index()

# Crear un grfico de rea para mostrar los promedios mensuales
plt.figure(figsize=(14, 7))
plt.fill_between(monthly_visibility['YearMonth'], monthly_visibility['visibility'], color="skyblue", alpha=0.4)
plt.plot(monthly_visibility['YearMonth'], monthly_visibility['visibility'], color="Slateblue", alpha=0.6, linewidth=2)
plt.title('Cambios en la Visibilidad Media Mensual')
plt.xlabel('Mes (Ao-Mes)')
plt.ylabel('Visibilidad Media (metros)')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mejor legibilidad
plt.grid(True)
plt.tight_layout()  # Ajustar automticamente los parmetros del subplot para dar espacio a las etiquetas
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
data = pd.read_csv('GCLP_All.csv')

# Convertir la columna 'dateTime' a datetime para facilitar el anlisis temporal
data['dateTime'] = pd.to_datetime(data['dateTime'])

# Filtrar para asegurarse de que solo incluimos entradas con visibilidad disponible
visibility_data = data[data['visibility'].notnull()]

# Ordenar los datos por fecha, si no estn ya ordenados
visibility_data = visibility_data.sort_values(by='dateTime')

# Crear el grfico de lneas
plt.figure(figsize=(14, 7))
plt.plot(visibility_data['dateTime'], visibility_data['visibility'], marker='o', linestyle='-', markersize=2, color='blue')
plt.title('Cambio de la Visibilidad a lo Largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Visibilidad (metros)')
plt.grid(True)
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mejorar legibilidad
plt.tight_layout()  # Ajusta la configuracin del layout para evitar cortes de etiquetas

# Mostrar el grfico
plt.show()
```
